## Color filaments by branch order

This shows how easy it is to do something that sounds difficult -- if you know how Geometry objects work. If you don't, I have another example in pyramidal_Neuron.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pyramidal_readExportedGeometry import *
import sys, os

```

```python
# Load the hocfiles
hocdir = '/home/alex/data/adriane/skeletons/hoc/'
hocfiles = [hocdir+h for h in os.listdir(hocdir) if h.split('.')[-1] == 'hoc']
geofiles = [demoReadsilent(h) for h in hocfiles]
# And pick a sample geofile for prototyping
geo = geofiles[0]
```

```python
# Get the branch orders and assign float values
geo.calcBranchOrder()
# branch orders
borders = [b.branchOrder for b in geo.branches]
maxorder = max(borders)
bfloats = [float(i)/maxorder for i in borders]
bcolors = plt.cm.jet(bfloats)
cmap = plt.cm.jet # Set colormaps (the only thing matlab makes easy)
segdict = {s.name: s for s in geo.segments} # Makes indexing easier
```

```python
# Iterate through the branches and color the segments by branchOrder
for b in range(len(geo.branches)):
    for s in range(len(geo.branches[b].nodes)-1):
        n0, n1 = geo.branches[b].nodes[s], geo.branches[b].nodes[s+1]
        plt.plot([n0.x, n1.x], [n0.y, n1.y], color=bcolors[b], 
                 alpha=0.4)
        
```

```python
# Wrap this in a function that takes a geo object
def plot_bOrder(geo, tofile=None, retax=False, maxorder=None):
    """ Plot a geometry file, color by branch order. """
    geo.calcBranchOrder()
    # branch orders
    borders = [b.branchOrder for b in geo.branches]
    if maxorder is None:
        maxorder = max(borders)
    bfloats = [float(i)/maxorder for i in borders]
    bcolors = plt.cm.jet(bfloats)
    cmap = plt.cm.jet
    for b in range(len(geo.branches)):
        for s in range(len(geo.branches[b].nodes)-1):
            n0, n1 = geo.branches[b].nodes[s], geo.branches[b].nodes[s+1]
            plt.plot([n0.x, n1.x], [n0.y, n1.y], color=bcolors[b], 
                     alpha=0.4)
    plt.axis('off')
    plt.title(geo.name)
    # color bar
    sc = []
    sc.append(plt.scatter([0,0], [0,0], c=[0., 1.], s=0.1,
                          vmin=0, vmax=maxorder, cmap=cmap))
    plt.colorbar(sc[-1])
    if tofile:
        plt.savefig(tofile)
        plt.close()
    if retax:
        return
    
    
    plt.show(); return
```

```python
# Let's make sure they all plot first
maxorders = []
for g in geofiles:
    g.calcBranchOrder()
    maxorders.append(max([b.branchOrder for b in g.branches]))
for g in geofiles:
    plot_bOrder(g, maxorder=max(maxorders))
```

```python
# Create a save list
outdir = '/home/alex/data/adriane/images/skeletons_border/'
for g in geofiles:
    plot_bOrder(g, maxorder=max(maxorders), tofile=outdir+g.name+'.eps')
```
