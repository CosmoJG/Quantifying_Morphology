## Tip distances vs their branch orders

Determine whether a branch order is related to its distance from the soma -- especially if it's a tip.


```python
# Imports
import sys
sys.path.append('../python/')
from neuron_readExportedGeometry import *
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# Load the sample neuron
geo = demoReadsilent('/home/alex/pyramidal_Neurons/878_067_GM_(Adriane)_scaled.hoc') # Change this
```

```python
# Find the tips
tipInds, tipLocs = geo.getTipIndices()
tipsegs = [seg for seg in geo.segments if seg.filamentIndex in tipInds]
for i in range(10):
    print('Segment %i (filamentIndex %i) has tip at %i'
          %(tipInds[i], tipsegs[i].filamentIndex, tipLocs[i]))
```

```python
# Now get the branch orders
geo.calcBranchOrder()
# Need to find which tips belong to which branch
tipNames = [tip.name for tip in tipsegs]
tipBranches = []
for t in tipNames:
    for b in geo.branches:    
        if t in b.tags:
            tipBranches.append(b)

if len(tipBranches) != len(list(set(tipBranches))):
    print('Somhow have %i branches, only want %i' 
          %(len(tipBranches),len(list(set(tipBranches)))))

# If needed -- might mess up the ordering of the tips...
#tipBranches = list(set(tipBranches))
#print('Found %i (unique) branches; have %i tips' %(len(tipBranches), len(tipsegs)))
```

```python
# Get the path distances for each tip
# Make a PathDistanceFinder object -- need to use distanceTo()
pDF = PathDistanceFinder(geo, geo.soma)
tipDists = [pDF.distanceTo(tipsegs[t], tipLocs[t]) for t in range(len(tipsegs))]
tipOrders = [b.branchOrder for b in tipBranches]
```

```python
# Plot the results
for t in range(len(tipsegs)):
    plt.plot(tipDists[t], tipOrders[t], 'o', color='blue', alpha=0.5)
plt.xlabel('Distance from soma (um)')
plt.ylabel('Order of tip branch')
```

```python
# Jason wants to color by max tortuosity
# Get the path tortuosity
# euclidean distance
def euc(pt0, pt1):
    return np.sqrt(sum([(pt1[i]-pt0[i])**2 for i in range(3)]))

somacoord = geo.soma.coordAt(0)
tipCoords = [tipsegs[t].coordAt(tipLocs[t]) for t in range(len(tipsegs))]
tipEucs = [euc(somacoord, tipCoords[t]) for t in range(len(tipsegs))]
tipTorts = [tipDists[t]/tipEucs[t] for t in range(len(tipsegs))]
maxTorts = max(tipTorts)
tfloats = [float(i)/maxTorts for i in tipTorts]
tcolors = plt.cm.jet(tfloats)
cmap = plt.cm.jet # Set colormaps (the only thing matlab makes easy)

# Plot it
for t in range(len(tipsegs)):
    plt.plot(tipDists[t], tipOrders[t], 'o', 
             color=tcolors[t], alpha=0.5, markeredgecolor='none')
plt.xlabel('Distance from soma (um)')
plt.ylabel('Order of tip branch')
plt.xlim([10,1000])
sc = []
sc.append(plt.scatter([0,0], [0,0], c=[0., 1.], s=0.1,
                      vmin=0, vmax=maxorder, cmap=cmap))
plt.colorbar(sc[-1])
```
